6. Explain the concept of generators in Python and how they are defined.
Generators in Python are functions that use yield to produce values lazily, one at a time, making them memory-efficient and suitable for large datasets.

7. What are the advantages of using generators over regular functions?
Generators are more memory-efficient and faster than regular functions because they produce values one at a time, without storing the entire sequence in memory.

8. What is a lambda function in Python and when is it typically used?
A lambda function in Python is a small, anonymous function defined using the lambda keyword, typically used for short-term, throwaway functions or when you need a simple function as an argument (e.g., in map(), filter(), or sorted()).

9. Explain the purpose and usage of the map() function in Python.
The map() function applies a given function to each item in an iterable and returns an iterator with the transformed results.

10. What is the difference between map(), reduce(), and filter() functions in Python?.
map() transforms elements, reduce() accumulates them into a single value, and filter() filters elements based on a condition.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def Total_Marks(marks):
    Sum = 0
    for mark in marks:
        Sum += mark
    return Sum
marks = [ 91, 86, 87 ,94 ,78 ]
result = Total_Marks(marks)
print(result)
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    return [num ** 2 for num in numbers]
numbers = [1, 2, 3, 4, 5]
result = square_numbers(numbers)
print(result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num)
```

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __iter__(self):
        self.a, self.b = 0, 1
        self.count = 0
        return self
    def __next__(self):
        if self.count < 10:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration
fib = FibonacciIterator()
for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_two(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines_from_list(lines):
    for line in lines:
        yield line

# Example usage with a predefined list of lines
lines = [
    "Hello, this is line 1.",
    "This is line 2.",
    "And here's line 3."
]

# Iterate over the generator to print each line
for line in read_lines_from_list(lines):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 3), (4, 1), (5, 2), (2, 4)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temperatures = [0, 10, 20, 30, 40, 50]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda x: x not in vowels, s))
string = "Hello World"
result = remove_vowels(string)
print(result)
```
